```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
data = pd.read_csv('world-happiness-report-2021.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data_columns = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
data = data[data_columns].copy()
```

```python
data.head()
```

```python
data.columns = ['country_name','regional_indicator','ladder_score','logged_GDP_per_capita','social_support','healthy_life_expectancy','freedom_to_make_life_choices','generosity','perceptions_of_corruption']
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# plot between happiness and gdp
```

```python
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Plot between Happiness score and GDP')
sns.scatterplot(x = data.ladder_score, y = data.logged_GDP_per_capita,hue = data.regional_indicator, s = 200);

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Ladder score')
plt.ylabel('Logged GDP per capita')
```

In the above graph we can observe the countries from western Europe highest happiness score and GDP per capita is also highest.
And the Afganistan has lowest happiness score according to 2021 world happiness report data. 

```python
# now we will plot pie plot to understand the GDP by region so by this we can know which has the highest contribution in world's GDP
```

```python
gdp_region = data.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
print(gdp_region)
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP by Region')
plt.ylabel('')
```

In the above pie plot we can observe the Sub-Saharan Africa and Western Europe are contributing highest GDP where as North America and ANZ are contributing less to the world GDP

```python
# now lets count number of countries in each region by using groupby function 
```

```python
total_country = data.groupby('regional_indicator')[['country_name']].count()
print(total_country)
```

```python
# lets plot correlation map to understand the relation between the features
```

```python
cor = data.corr(method = 'pearson')
f, ax = plt.subplots(figsize = (10,5))
sns.heatmap(cor, mask = np.zeros_like(cor, dtype=np.bool),
           cmap = 'Blues', square = True, ax = ax)
```

In the above heatmap the dark blue region indicates the highest correlation and the white grey color indicates the low correlation.
Here we can observe there is very low or negative correlation between happiness_score and the percepton of curruption bcz if the citizens of the country feel that there is lot of curruption in the country then the obviously the happiness score would low.


```python
# corruption in regions

curruption = data.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
curruption
```

```python
# from the above table we can observe the North America and ANZ and Western Europe has the lowest curruption
```

```python
# so now lets visualize this data in the pieplot
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Perception of curruption in various regions')
plt.xlabel('Regions', fontsize = 15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(curruption.index, curruption.perceptions_of_corruption)
```

```python
# now lets fetch out the top 10 high life expectancy countries and bottom 10 and visualize them
```

```python
top_10 = data.head(10)
bottom_10 = data.tail(10)  
# ha is horizontal allignment
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,6))
plt.tight_layout(pad = 2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels , rotation=45, ha = 'right')
sns.barplot(x = top_10.country_name, y = top_10.healthy_life_expectancy, ax = axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')


xlabels = bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels , rotation=45, ha = 'right')
sns.barplot(x = bottom_10.country_name, y = bottom_10.healthy_life_expectancy, ax = axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')
```

so from the above barplot we can clearly visualize that the happiest countries have the life expectancy above 70 years where as the bottom 10 least happy counties have the life expectancy of less than 60 years.
so if we are from the Top 10 happiest counties our life expectancy is 10 years more than the least happy countries.

```python
# now lets plot the graph between freedom to make life choices and the happiness score
```

```python
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x = data.freedom_to_make_life_choices, y = data.ladder_score, hue = data.regional_indicator, s = 200)
plt.legend(loc = 'upper left', fontsize = '12')
plt.xlabel('Freedom to Make Choices')
plt.ylabel('Happiness Score')
```

So here in the western countries the freedom to make life choices is higher so the happiness score is also high. And in the middle east africa the freedom to make life choices is lower so the happiness score in that region is also low.

In the Southeast Asian countries eventhough the happiness score is comparitively lower but the freedom to make life choices is more than 0.8.

and in South Asia i.e Afganisthan the happiness score as well as the freedom to make life choices are lower.



```python
# lets know top 10 corruption countries
```

```python
# now lets visualize the countries with highest perception of corruption 
```
